```python
# Removing duplicate rows from the DataFrame
covid.drop_duplicates(inplace=True)
```

```python
# Dropping irrelevant columns from the dataset
covid.drop(['USMER'], inplace =True, axis=1)
covid
```

```python
print(covid.columns)
```

```python
# Renaming columns for better clarity and consistency
covid.rename(columns=
             {'PATIENT_TYPE':'HOSPITALIZED',
              'DATE_DIED':'DEAD',
              'INMSUPR': 'IMMUNOSUPPRESSED',
              'CLASIFFICATION_FINAL': 'COVID_TEST_RESULT',
              'TOBACCO': "SMOKER",
              'CARDIOVASCULAR': 'CARDIOVASCULAR_DISEASE',
              'RENAL_CHRONIC': 'CHRONIC_KIDNEY_DISEASE',
              'HIPERTENSION' : 'HYPERTENSION',
              'ICU': 'ADMITTED_TO_ICU'}, inplace=True)
covid
```

```python
# Reordering columns to improve dataset structure
covid = covid.loc[:,
                  ['COVID_TEST_RESULT', 
                   'HOSPITALIZED',
                   'INTUBED',
                   'ADMITTED_TO_ICU',
                   'DEAD',
                   'AGE',
                   'SEX',
                   'PREGNANT',
                   'SMOKER',
                   'PNEUMONIA',
                   'COPD',
                   'ASTHMA',
                   'IMMUNOSUPPRESSED',
                   'CARDIOVASCULAR_DISEASE',
                   'HYPERTENSION',
                   'CHRONIC_KIDNEY_DISEASE',
                   'DIABETES',
                   'OBESITY',
                   'OTHER_DISEASE',
                   'MEDICAL_UNIT']]
```

```python
covid
```

```python
# Replacing death date with 1 (True) for deceased, '9999-99-99' with 0 (False) for alive
covid['DEAD'] = covid['DEAD'].replace({'9999-99-99': 0})
covid['DEAD'] = covid['DEAD'].apply(lambda x: 1 if x != 0 else 0)
covid
```

```python
# Replacing numeric gender values with 'female' and 'male'
covid['SEX'] = covid['SEX'].replace({1: 'female', 2: 'male'})
covid
```

```python
# Checking unique values and their count for each column
for column_title in covid.columns:
    print(column_title, covid[column_title].unique(), covid[column_title].nunique())
```

```python
# Replacing 2 with 0 ('False'), and 97, 98, 99 with 'NaN' for missing values. 'NaN' values are excluded in plots.
covid[['INTUBED','ADMITTED_TO_ICU','PREGNANT','SMOKER','PNEUMONIA','COPD','ASTHMA','IMMUNOSUPPRESSED',
       'CARDIOVASCULAR_DISEASE','HYPERTENSION','CHRONIC_KIDNEY_DISEASE', 'DIABETES','OBESITY','OTHER_DISEASE']] = covid[[
           'INTUBED','ADMITTED_TO_ICU','PREGNANT','SMOKER','PNEUMONIA','COPD','ASTHMA','IMMUNOSUPPRESSED',
           'CARDIOVASCULAR_DISEASE','HYPERTENSION', 'CHRONIC_KIDNEY_DISEASE','DIABETES','OBESITY','OTHER_DISEASE']].replace({2: 0, 97: 'NaN', 98: 'NaN', 99: 'NaN'})
```

```python
# In this column, 1 and 2 represent the opposite meaning, so replacing 1 with 0 and 2 with 1
covid['HOSPITALIZED'] = covid['HOSPITALIZED'].replace({1: 0, 2: 1})
covid
```

```python
# Converting 'COVID_TEST_RESULT' values: 1, 2, or 3 to 1 (positive), all others to 0 (negative)
covid['COVID_TEST_RESULT'] = covid['COVID_TEST_RESULT'].apply(lambda x: 1 if x in [1,2,3] else 0)
covid.sample(30)
```

```python
# Displaying unique values and the count of unique entries for each column in the dataset
for column_title in covid.columns:
    print(column_title, covid[column_title].unique(), covid[column_title].nunique())
```
